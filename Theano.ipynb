```python
def animation_for_model(model, initial_state, time_horizon, path=None):
    problem = DoublePendulumProblem()
    plant = problem.plant()
    plant.controller = model.controller()
    #plant.controller = lambda x,t : 0.0
    traj = problem.compute_trajectory(plant, initial_state, time_horizon)
    problem.plot_u(plant, traj)
    print('Sampled trajectory of cost %f.' % (problem.cost(traj),))
    animation = problem.animate_trajectory(plant, traj)
    if path is not None:
        animation.save(path, fps=20)
        animation._save_path = path
    return animation
```

```python
model = temporal_model

goal_state = upside = np.array([pi,0,pi,0])

def perturb_state(state):
    return state + np.array([
            np.random.uniform(- pi/2, pi/2),
            np.random.standard_normal(),
            np.random.uniform(-pi/2, pi/2),
            np.random.standard_normal()
        ])* (np.random.rand(4)> 0.1)

timesteps = 20

test_bundle = []
def score_model(model):
    TEST_SIZE = 100
    if len(test_bundle) == 0:
        for _ in range(TEST_SIZE):
            test_bundle.append(perturb_state(goal_state))
    score = 0.0
    for test in test_bundle:
        score += model.cost_fun['test'](test, goal_state, timesteps)
    return score / TEST_SIZE

def test_model(model, lr):
    model.reset_weights()
    model.set_learning_rate(lr)
    print(model.predict['test'](upside))
    ITERS = 30000
    for itera in range(ITERS):
        initial_state = perturb_state(upside)
        cost = model.update_fun(initial_state, goal_state, timesteps)
        #if lr < 0.0005:
        #    break
        if itera % 500 == 0:
            clear_output(wait=True)
            print('Iteration %d (learning rate: %f): %f'  % (itera, lr,  score_model(model)))
        if itera > 0 and itera % (ITERS/10) == 0:
            lr = lr * 0.8
            model.set_learning_rate(lr)
        
    print(model.predict['test'](upside))

test_model(model, 0.01)
```

```python
def show_model(theta_1 = pi, theta_1_dot = 0., theta_2 = pi, theta_2_dot = .5, horizon = 1000):
    obs = np.zeros([horizon, plant.num_states])

    obs[0,:] = np.array([theta_1, theta_1_dot, theta_2 , theta_2_dot])

    for i in np.arange(1, horizon):
        obs[i,:] = model.teleportation['test'](obs[i-1,:])[0]
        obs[i,0] %= 2*pi
        obs[i,2] %= 2*pi


    plt.plot(np.linspace(0, horizon * plant.dt, horizon), obs)
    plt.legend([r"$\theta_1$", r"$\dot{\theta}_1$", r"$\theta_2$", r"$\dot{\theta}_2$"]);

interact(show_model,
         horizon=IntSliderWidget(min=1, max=2000, value=100),
         theta_1_dot=FloatSliderWidget(min=-10, max=10, value=0.0),
         theta_2_dot=FloatSliderWidget(min=-10, max=10, value=0.0),
         theta_2 = FloatSliderWidget(min=-2*pi, max=2*pi, value=pi),
         theta_1 = FloatSliderWidget(min=-2*pi, max=2*pi, value=pi))
```

### Sanity Check:
dynamics between Theano and Scipy agree

```python
u = T.vector()
next_pos = theano.function([model._x,u], model.dynamics(model._x, u), allow_input_downcast=True)
for _ in range(1000):
    #state = np.array([pi, 0., pi, 0.])
    #u = np.array([0.0])
    state = np.random.standard_normal([4])* 4
    u = np.random.standard_normal([1])
    
    theano_thinks = next_pos(state, u) # theano
    theano_thinks = theano_thinks[0, :]
    theano_thinks[0] %= 2*pi
    theano_thinks[2] %= 2*pi
    plant.controller = lambda x,t: u
    simon_thinks = state + plant.simulation_derivs(state, 0) * plant.dt
    
    error = np.linalg.norm(theano_thinks-simon_thinks)
    if error > 1e-3:
        print(theano_thinks)
        print(simon_thinks)
        print(error)
```
